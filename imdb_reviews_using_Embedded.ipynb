```python
from keras.preprocessing.text import one_hot
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.embeddings import Embedding
```

```python
docs = df['Reviews']
labels = df['Status']
```

```python
docs
```

```python
labels
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest = train_test_split(docs,labels,test_size = 0.3,random_state = 42)
```

```python
print(Xtrain.shape)
print(Xtest.shape)
print(Ytrain.shape)
print(Ytest.shape)
```

```python
vocab_size = 500 
```

# The one_hot function expects the following arguments:

```python
Xtrain = [one_hot(d,vocab_size,filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~',lower = True,split = ' ')for d in Xtrain]
Xtest = [one_hot(d,vocab_size,filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~',lower = True,split = ' ')for d in Xtest]
```

```python
Xtrain
```

```python
Xtest
```

# Pad_sequence

```python
max_length = 25
Xtrain = pad_sequences(Xtrain,maxlen = max_length,padding = 'pre')
Xtest = pad_sequences(Xtest,maxlen = max_length,padding = 'pre')
```

```python
Xtrain[7],Xtest[7]
```

# Creating Model

```python
from keras.layers import Dropout
```

```python
model = Sequential()
model.add(Embedding(vocab_size,8,input_length = max_length))
model.add(Dropout(0.3))
model.add(Flatten())
model.add(Dense(1,activation = 'sigmoid'))
model.compile(optimizer = 'adam',loss = 'binary_crossentropy',metrics = ['acc'])
```

```python
model.summary()
```

```python
model.fit(Xtrain,Ytrain,epochs = 20,verbose = 0)
```

```python
#short form 
#model.fit(Xtrain,Ytrain,validation_data=(Xtest,Ytest),epochs=20,batch_size = 64)
```

# Training

```python
loss,accuracy= model.evaluate(Xtrain,Ytrain,verbose = 1)
print('Loss : ',loss)
print("Accuracy : ",accuracy)
```

# Testing

```python
loss1,accuracy1 = model.evaluate(Xtest,Ytest,verbose = 1)
print('Loss : ',loss1)
print("Accuracy : ",accuracy1)
```

```python
predict = model.predict(Xtest)
print(predict)
```

```python
Ypred = np.round(predict) 
print(Ypred)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Ytest,Ypred))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Ytest,Ypred))
```
